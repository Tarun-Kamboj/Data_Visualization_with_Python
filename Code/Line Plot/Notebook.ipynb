# Line Plots
---

Let's import all the dependencies first.

```python
# primary data structure library
import pandas as pd       

# primary plotting library
import matplotlib as mpl       

# importing the pyplot layer of matplotlib for easy usage
import matplotlib.pyplot as plt        

# optional: for ggplot-like style of plots
mpl.style.use(['ggplot'])        

# using the inline backend
%matplotlib inline 
```

Now we'll create the DataFrame.

```python
# dict for dummy data to be used in dataframe
data = {
    'India': [190, 200, 220, 390, 1160, 1600],  
    'China': [320, 400, 390, 580, 1340, 1390]
}       

# list for indexes to be used in dataframe
index = [1800, 1850, 1900, 1950, 2000, 2050]        

# making dataframe
df = pd.DataFrame(data, index=index)      

# showing dataframe
df.T   
```

Let's draw a line plot for single column or series.

```python
# plotting for single column and specifying the size and kind of plot
df['India'].plot(kind='line', figsize=(10, 6))  

# giving titles and labels
plt.title('Population of India')       
plt.ylabel('In Millions')    
plt.xlabel('Years')     

 # showing the plot
plt.show()         
```

We can plot the graph for the whole dataframs with each column having a line in the line graph.

```python
# plotting for whole dataframe 
df.plot(kind='line', figsize=(10, 6))        
plt.title('Population of India and China')
plt.ylabel('In Millions')
plt.xlabel('Years')
plt.show() 
```

We can even annotate a point and write text on the graph.

```python
df.plot(kind='line', figsize=(10, 6))
plt.title('Population of India and China')
plt.ylabel('In Millions')
plt.xlabel('Years')

 # annotating a point in the plot, syntax: plt.text(x, y, "text") 
plt.text(1985, 1400, "India will surpass china \naround year 2024")    
plt.show() 
```
