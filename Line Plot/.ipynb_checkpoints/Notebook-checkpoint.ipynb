# Line Plots
---

## Introduction

A `line chart` or l`ine plot` is a type of plot which displays information as a series of data points called `markers` connected by straight line segments. It is a basic type of chart common in many fields.
Use line plot when you have a continuous data set. These are best suited for trend-based visualizations of data over a period of time.

Let's import all the dependencies first.

```python
import pandas as pd       # primary data structure library
import matplotlib as mpl       # primary plotting library
import matplotlib.pyplot as plt        # importing the pyplot layer of matplotlib for easy usage
mpl.style.use(['ggplot'])        # optional: for ggplot-like style of plots
%matplotlib inline 
# using the inline backend
```

Now we'll create the DataFrame.

```python
data = {'India': [190, 200, 220, 390, 1160, 1600],  'China': [320, 400, 390, 580, 1340, 1390]}       # dict for the data to be used in dataframe
index = [1800, 1850, 1900, 1950, 2000, 2050]        # list for indexes to be used in dataframe
df = pd.DataFrame(data, index=index)      # making dataframe
df      # showing dataframe
```

Let's draw a line plot for single column or series.

```python
df['India'].plot(kind='line', figsize=(14, 8))      # plotting for single column and specifying the size and kind of plot
plt.title('Population of India')       # title of the plot
plt.ylabel('In Millions')          # label for the y-axis
plt.xlabel('Years')          # label for the x-axis
plt.show()          # showing the plot
```

```python
df.plot(kind='line', figsize=(14, 8))        # plotting for whole dataframe and specifying the size and kind of plot
plt.title('Population of India and China')
plt.ylabel('In Millions')
plt.xlabel('Years')
plt.show() 
```

```python
df.plot(kind='line', figsize=(14, 8))
plt.title('Population of India and China')
plt.ylabel('In Millions')
plt.xlabel('Years')
plt.text(1985, 1400, "India will surpass china \naround year 2024")     # annotating a point in the plot, syntax: plt.text(x, y, "text") 
plt.show() 
```
